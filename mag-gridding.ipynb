```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Step 1: Read the Excel file containing magnetometry data
file_path = r'D:\University & Work\Datasets\geophysical data\Mag and Grav\morvarid mine\Data\Mag-morvarid.xlsx'
data = pd.read_excel(file_path)

# Step 2: Extract x, y, z, and mag columns
x = data['x']
y = data['y']
z = data['topo']
mag = data['mag']

# Step 3: Perform minimum curvature interpolation
# Define a grid for the interpolation
grid_x, grid_y = np.meshgrid(np.linspace(min(x), max(x), 100), np.linspace(min(y), max(y), 100))

# Interpolate the magnetic data using minimum curvature
grid_mag = griddata((x, y), mag, (grid_x, grid_y), method='cubic')

# Step 4: Create a map with a color legend
plt.figure(figsize=(10, 8))
plt.imshow(grid_mag, extent=(min(x), max(x), min(y), max(y)), origin='lower', cmap='viridis')
plt.colorbar(label='Magnetic Field (nT)')
plt.savefig('original_magnetic_field.png', dpi=300)  # Change the format and DPI as needed
```
